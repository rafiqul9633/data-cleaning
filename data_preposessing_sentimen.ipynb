```python
import neattext.functions as nfx
```

```python
dt["text"].iloc[3]
```

# extracted hashtag dalam text tweet

```python
dt["text"].apply(nfx.extract_hashtags)
```

# simpan extracted hashtag dalam kolom terbaru

```python
dt["extracted_hashtag"] = dt["text"].apply(nfx.extract_hashtags)
```

```python
dt[["extracted_hashtag","hashtags"]]
```

```python
dt = dt.drop("extracted_hashtag", axis=1)
```

# extract dan simpan user name dalam text tweet ke kolom baru

```python
dt["extracted_user_mention"] = dt["text"].apply(nfx.extract_userhandles)
```

```python
dt["extracted_user_mention"]
```

# hapus hashtag dalam text tweet

```python
dt["clean_tweets"] = dt["text"].apply(nfx.remove_hashtags)
```

# hapus user_name dalam text tweet

```python
dt["clean_tweets"] = dt["clean_tweets"].apply(lambda x: nfx.remove_userhandles(x))
```

```python
dt["clean_tweets"].iloc[10]
```

# hapus double space dalam text tweet

```python
dt["clean_tweets"] = dt["clean_tweets"].apply(nfx.remove_multiple_spaces)
```

# hapus url link dalam text tweet

```python
dt["clean_tweets"] = dt["clean_tweets"].apply(nfx.remove_urls)
```

# hapus emoji dalam text tweet

```python
dt["clean_tweets"] = dt["clean_tweets"].apply(nfx.remove_emojis)
```

# hapus tanda baca dalam text tweet

```python
dt["clean_tweets"] = dt["clean_tweets"].apply(nfx.remove_puncts)
```

```python
dt["clean_tweets"] = dt["clean_tweets"].apply(nfx.remove_special_characters)
```

```python
dt["clean_tweets"] = dt["clean_tweets"].apply(nfx.remove_currency_symbols)
```

# hapus tanda angka dalam text tweet

```python
dt["clean_tweets"] = dt["clean_tweets"].apply(nfx.remove_numbers)
```

# mensortir tweet yang kosong

```python
dt.sort_values("clean_tweets", inplace = True)
```

```python
dt
```

# hapus duplicate

```python
dt.drop_duplicates(subset = "clean_tweets", keep = "first", inplace = True)
```
